```python
#19. Calculate the sum of the numbers from 1 to 100
# Write your code here

import functools

my_numbers = []
for i in range(1, 101):
    my_numbers.append(i)

#print(my_numbers)    
    
sum_reduce = functools.reduce(lambda x, y : x + y, my_numbers)
print("The sum of the numbers from 1 to 100  ", sum_reduce)

```
